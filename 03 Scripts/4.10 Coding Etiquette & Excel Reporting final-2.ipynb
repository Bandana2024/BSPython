## Creating column base on age
Creating age group
18-30 as Young adult
65 and older as Elderly
or else = Adult

```python
#If the user is younger than 30, then user will be labeled as “Young adult” using .loc
Visual_data.loc[Visual_data['Age'] < 30, 'age_group'] = 'Young adult'
```

```python
Visual_data.loc[(Visual_data['Age'] >= 30) & (Visual_data['Age'] < 65),'age_group'] = 'Adult'
```

```python
Visual_data.loc[Visual_data['Age'] >= 65, 'age_group'] = 'Elderly'
```

Creating coulmn base on income 
Income groups will have 3 categories
if income less than 60K will be addess as "Low income'
if income higher than 250K will be addess as "High income'
rest income that fall inbtween will be ddressed as 'Middle income'

```python
result = []

for value in Visual_data["income"]:
  if value < 60000:
    result.append("Low income")
  elif value >= 250000:
    result.append("High income")
  else:
    result.append("Middle income")
```

```python
result
```

```python
Visual_data['income_Group'] = result
```
