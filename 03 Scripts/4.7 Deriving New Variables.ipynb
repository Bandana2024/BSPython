```python
#defining function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#where the lable be returned : apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#for_loop function for creating a busiest day 
result = [] #his will act as the empty shell, into which you can place the results from your loop

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#create a new column within your ords_prods_merge dataframe and set it equal to result
ords_prods_merge['busiest_day']=result
```

#task
theyd also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# creating a new column using loop funtion
X_result = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        X_result.append("Busiest days")
    elif value == 4 or value == 3:
        X_result.append("Slowest days")
    else:
        X_result.append("Regularly days")
```

```python
X_result
```

```python
#create a new column within your ords_prods_merge dataframe and set it equal to X_result
ords_prods_merge['busiest_days']=X_result
```

```python
#checking the fequency of the result
ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# creating a new column with most orders, average orders and fewest orders using loop funtion
Y_result=[]
for value in ords_prods_merge['order_hour_of_day']:
    if value == 10:
        Y_result.append("Most orders")
    elif value == 3:
        Y_result.append("Fewest orders")
    else:
        Y_result.append("Average orders")
```

```python
#create a new column within your ords_prods_merge dataframe and set it equal to Y_result
ords_prods_merge['busiest_period_of_day']=Y_result
```
