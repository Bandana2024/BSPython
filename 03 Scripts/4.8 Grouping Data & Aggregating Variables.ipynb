## Task

The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
#grouping and using aggregate
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The result actually shows that the mean of new customer are infact hight than of regular and loyal customer. 

This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders.

```python
#first create a mean_buyer (using aggregation function)
ords_prods_merge['mean_buyer'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#now create new column for spender using .loc function
ords_prods_merge.loc[ords_prods_merge['mean_buyer']<10,'Spending_flag']= 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_buyer']>=10,'Spending_flag']= 'Higher spender'
```

```python
ords_prods_merge['Spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head()
```

the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
f the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#creating a median value column
ords_prods_merge['median_ordering'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#creating frequency flag column using loc function
ords_prods_merge.loc[ords_prods_merge['median_ordering']>20,'Order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_ordering']>10) & (ords_prods_merge['median_ordering']<=20),'Order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_ordering']<=10,'Order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['Order_frequency_flag'].value_counts(dropna=False)
```
